```python
from __future__ import print_function
import numpy as np
from time import time # Để so sánh thời gian chạy
d, N = 1000, 10000 # Kích thước, số lượng điểm đào tạo
```

```python
X = np.random.randn(N, d)
print(X)
```

```python
z = np.random.randn(d)
print(z)
```

```python
#Hàm này tính khoảng cách Euclid giữa 2 vector z và x. Tính hiệu z - x rồi
# lấy bình phương l2 norm của nó.
def dist_pp(z, x):
    d = z - x.reshape(z.shape) # Chuyển đổi để x và z có cùng kích thước
    return np.sum(d*d)
```

```python
#Hàm này tính bình phương khoảng cách giữa z và mỗi hàng của ma trận X.
#Trong hàm này, các khoảng cách được xây dựng dựa trên việc tính từng
#giá trị dist_pp(z, X[i])
def dist_ps_naive(z, X):
    N = X.shape[0]
    res = np.zeros((1, N))
    for i in range(N):
        res[0][i] = dist_pp(z, X[i])
    return res
```

```python
#Hàm này tính bình phương khoảng cách giữa z và mỗi hàng của x. Ta cần
#tính tổng bình phương các phần tử của mỗi điểm dữ liệu trong X và tính 
#X.dot(z)
def dist_ps_fast(z, X):
    X2 = np.sum(X*X, 1)
    z2 = np.sum(z*z)
    return X2 + z2 - 2*X.dot(z)
```

```python
t1 = time()
print(t1)
```

```python
D1 = dist_ps_naive(z, X)
print(D1)
print('naive point2set, running time', time() - t1, 's')
```

```python
t1 = time()
D2 = dist_ps_fast(z, X)
print(D2)
print('fast point2set, running time:', time() - t1, 's')
print('Result difference:', np.linalg.norm(D1 - D2))
```
